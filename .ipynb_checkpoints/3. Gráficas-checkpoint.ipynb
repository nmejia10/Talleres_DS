## 1. Introducción

La visualización de datos es la representación gráfica de los datos. Implica producir imágenes que comuniquen las relaciones entre los datos representados a los espectadores.

La visualización de datos es una parte esencial del análisis de datos y el aprendizaje automático. Usaremos bibliotecas de Python como Matplotlib y Plotly para aprender y aplicar algunas técnicas populares de visualización de datos.

## 2. Taller

Como en casos anteriores, lo primero que debemos hacer es importar las librerías con las que trabajaremos. En este caso importemos `pandas`, `numpy`, `matplotlib`, `plotly`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

### 2.1 Gráfica de Línea

```python
x = np.arange(10)
y = x ** 2
```

En matplotlib podemos hacer una simple gráfica de linea con el comando `plot(x, y)`

```python
# Matplotlib
plt.plot(x, y);
```

En plotly es un poco mas complicado a primera vista. Debemos primero crear una figura (espacio donde haremos la gráfica) y luego espicificar el tipo de gráfica que queremos. En este caso `Scatter(x, y)` nos sirve para crear gráficos de lineas o de puntos, dependiendo de como especifiquemos el parámetro `mode`

```python
# Plotly
fig = go.Figure(data = go.Scatter(y = y, x = x))
fig.show()
```

Las gráficas no se ven mal, pero estan muy "vacias". Podemos modificar lo ejes, poner leyendas, títulos, cambiar los colores y más, así como lo haríamos en Excel.

Si quieren una referencia de los coores disponibles pueden encontrarla en:
* Colores Matplotlib: https://matplotlib.org/stable/gallery/color/named_colors.html
* Colores Plotly: https://plotly.com/python/builtin-colorscales/

```python
# Matplotlib
plt.plot(x, y, color = 'r')
plt.xlabel('Nombre Eje X')
plt.ylabel('Nombre Eje Y')
plt.title('Título Gráfica')
plt.legend(['Leyenda'])
plt.show();
```

```python
fig = go.Figure(data = go.Scatter(y = y, x = x, name = 'Leyenda', line = dict(color = 'red')))
fig.update_layout(title = 'Título Gráfica',
                    xaxis_title = 'Nombre Eje X',
                    yaxis_title = 'Nombre Eje Y',
                    showlegend = True)

fig.show()
```

Si quisieramos graficar varias lineas o series simplemente repetimos los comandos anteriores

```python
x2 = np.arange(10)
y2 = (x2 ** 2)[::-1]
```

```python
plt.plot(x, x**2, marker = '+')
plt.plot(x2, y2, marker = '.')
plt.xlabel('Nombre Eje X')
plt.ylabel('Nombre Eje Y')
plt.title('Título Gráfica')
plt.legend(['Leyenda1', 'Leyenda2'])
plt.show();
```

```python
fig = go.Figure(data = go.Scatter(y = y, x = x, name = 'Leyenda1'))
fig.add_trace(go.Scatter(y = y2, x = x2, name = 'Leyenda2'))
fig.update_layout(title = 'Título Gráfica',
                    xaxis_title = 'Nombre Eje X',
                    yaxis_title = 'Nombre Eje Y',
                    showlegend = True)

fig.show()
```

### 2.2 Diagrama de Barras

Para la gráfica de barras utilizamos el comando `plt.bar(x, y)` en Matplolib y `go.Bar(x, y)` en Plotly.

```python
# Matplotlib
plt.bar(x, y);
```

```python
# Plotly
fig = go.Figure(data = go.Bar(y = y, x = x))
fig.show()
```

#### 2.2.1 Histograma

Un histograma es una representación gráfica de una variable en forma de barras, donde la superficie de cada barra es proporcional a la frecuencia de los valores representados. Sirven para obtener una "primera vista" general, o panorama, de la distribución de la población, o de la muestra, respecto a una característica, cuantitativa y continua (como la longitud o el peso)

```python
x = np.random.normal(5, 1.5, size=1000)
```

Para hacer un Histograma utilizamos el comando `plt.hist(x, bins)` en Matplolib y `go.Histogram(x, y)` en Plotly.

```python
# Matplotlib
plt.hist(x, bins = 25);
```

```python
fig = go.Figure(data = go.Histogram(x = x, nbinsx = 25))
fig.show()
```

### 2.3 Gráfica de Pastel

```python
labels = ['label 1', 'label 2', 'label 3', 'label 4']
values = [0.5, 0.25, 0.15, 0.1]
```

```python
# Matplotlib
plt.pie(values);
plt.legend(labels, loc = 'upper right');
```

```python
#Plotly
fig = go.Figure(data = go.Pie(labels = labels, values = values))
fig.show()
```

### 2.4 Gráfico de Cajas y Bigotes

```python
x = np.random.normal(5, 1.5, size=1000)
```

```python
# Matplotlib
plt.boxplot(x);
```

```python
#Plotly
fig = go.Figure(data = go.Box(y = x))
fig.show()
```

### 2.4.1 Gráfica de Violin

```python
# Matplotlib
plt.violinplot(x);
```

```python
#Plotly
fig = go.Figure(data = go.Violin(y = x, box_visible = True))
fig.show()
```
